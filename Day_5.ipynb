```python
# WAP to input list & an element & remove all occurrencies of given element from list.
L = list(map(int, input("Enter list elements separated by space: ").split()))
element = int(input("Enter the element to remove: "))

# Removing all occurrences
L = [x for x in L if x != element]

print("Updated list after removing all occurrences:", L)
```

```python
from datetime import datetime

class Workout:
    def __init__(self, date, exercise, duration, calories):
        self.date = date
        self.exercise = exercise
        self.duration = duration
        self.calories = calories

    def __str__(self):
        return f"{self.date}: {self.exercise} for {self.duration} minutes, {self.calories} calories burned"

class FitnessTracker:
    def __init__(self, user_name):
        self.user_name = user_name
        self.workouts = []

    def add_workout(self, date, exercise, duration, calories):
        self.workouts.append(Workout(date, exercise, duration, calories))
        print("Workout added successfully!")

    def view_workouts(self):
        if not self.workouts:
            print("No workouts recorded yet.")
        else:
            print(f"{self.user_name}'s Workouts:")
            for workout in self.workouts:
                print(workout)
    
    def save_data(self, filename="fitness_data.txt"):
        with open(filename, "w") as file:
            file.write(f"{self.user_name}\n")
            for workout in self.workouts:
                file.write(f"{workout.date},{workout.exercise},{workout.duration},{workout.calories}\n")
        print("Data saved successfully!")
    
    def load_data(self, filename="fitness_data.txt"):
        try:
            with open(filename, "r") as file:
                lines = file.readlines()
                self.user_name = lines[0].strip()
                self.workouts = [Workout(*line.strip().split(",")) for line in lines[1:]]
            print("Data loaded successfully!")
        except FileNotFoundError:
            print("No saved data found.")

def main():
    user_name = input("Enter your name: ")
    tracker = FitnessTracker(user_name)
    
    while True:
        print("\n1. Add Workout\n2. View Workouts\n3. Save Data\n4. Load Data\n5. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            date = input("Enter date (YYYY-MM-DD): ") or datetime.today().strftime('%Y-%m-%d')
            exercise = input("Enter exercise name: ")
            duration = input("Enter duration (minutes): ")
            calories = input("Enter calories burned: ")
            tracker.add_workout(date, exercise, duration, calories)
        
        elif choice == "2":
            tracker.view_workouts()
        
        elif choice == "3":
            tracker.save_data()
        
        elif choice == "4":
            tracker.load_data()
        
        elif choice == "5":
            print("Exiting...")
            break
        
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
